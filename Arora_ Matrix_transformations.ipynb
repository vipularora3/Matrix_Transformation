```python
# Create a black canvas and put the image there

canvas=np.zeros((500,500,3))
for i in range(250):
  for j in range(201):
    try:
        canvas[i+50,j+50]=im_array[i,j]
    except:
        pass


Image.fromarray(canvas.astype(np.uint8))
```

```python
# This is a rotation matrix
# https://en.wikipedia.org/wiki/Rotation_matrix
import math

def rot_mat(theta):
    return np.array([[math.cos(theta),-math.sin(theta)],[math.sin(theta),math.cos(theta)]])

# To transform (move) a point you dot-product the matrix times each point
def mat_t(mat,point):
  return np.dot(mat,point)
```

```python
# Now rotate the image using matrix transformations
import numpy as np

canvas=np.zeros((500,500,3))

angle=135

for i in range(250):
  for j in range(201):
    p=mat_t(rot_mat(angle),[i,j])
    canvas[int(p[0])+350,int(p[1])+350]=im_array[i,j]

Image.fromarray(canvas.astype(np.uint8))
```
